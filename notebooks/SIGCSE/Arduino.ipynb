```python
from Processing import *
window(800, 160)
background(0)
noStroke()
for x in range(800):
    y = analogRead(2)/5
    fill(255, 0, 196)
    ellipse(x, y, 3, 3)
    delay(10)
    calico.animate(get())
```

```python
window(800, 160)
background(0)
noStroke()
for x in range(800):
    y = analogRead(2)/5
    if y > 100:
        ledOn()
    else:
        ledOff()
    fill(255, 0, 196)
    ellipse(x, y, 3, 3)
    delay(10)
    calico.animate(get())
```
