```python
%%%lang fsharp
let cam = calico.makeCameraWidget()
calico.display(cam)
```

```python
let pic = Graphics.Picture(cam.imageurl)
for y in 0..pic.getHeight() do
    for x in 0..pic.getWidth() do
        let px = Graphics.getPixel(pic, x, y)
        px.setRed(255)
    done
    calico.animate(pic)
done
```
