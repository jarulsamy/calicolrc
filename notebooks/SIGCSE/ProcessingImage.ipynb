```python
cam = calico.makeCameraWidget()
calico.display(cam)
```

```python
from Processing import *

window(500, 300)
noStroke()
fill(196, 64)
background(24)
def draw(label, v):
    img = loadImage(cam.imageurl)
    gx = (v/100 * width())
    background(24)
    for i in range(0, int(gx), 10):
        fill(128, i)
        ellipse(i, 150, 25, 25)
    image(img, i, 150-96/2, 128, 96)
    calico.animate(get())

widget = calico.FloatSliderWidget()
calico.display(widget)
widget.on_value_change(draw, "value")
```
