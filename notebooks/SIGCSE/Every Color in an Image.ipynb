The challenge is to create a 256x192 image using each 15-bit color exactly once and only once. See the stack exchange site for a more detailed
[description](http://codegolf.stackexchange.com/questions/22144/images-with-all-colors). This notebook tackles the challenge using [Calico](http://calicoproject.org/) & [F#](http://fsharp.org/).

```python
%%%lang fsharp
```

First let's create an empty image that's 256 pixels by 128 pixels.

```python
let p  = Graphics.makePicture(256, 128)
p
```

This isn't necessary, but the image is white by default, so let's make all the pixels light gray so we can see the extent of the image.

```python
for px in p.getPixels() do
    px.setRGB(224, 224, 224)
p
```

```python
let pixels = p.getPixels()
```

```python
pixels |> Seq.length
```

Now the F# magic let's us create a sequence of all the colors (spaced by 8 so we only use 32 colors per channel, i.e. 5-bits)

```python
let colors = seq {for r in  0 ..8 .. 255 do 
                  for g in  0 ..8.. 255 do 
                  for b in  0 ..8.. 255 do yield (r,g,b) }
```

Let's check and make sure our sequence has 32 x 32 x 32 elemnts. (Using the nice F# pipe syntax)

```python
let ncolors =  colors |> Seq.length 
let npixels = pixels |> Seq.length
printfn "number of colors = %d, number of pixels = %d, expected = %d" ncolors npixels (32 * 32 * 32)
```

OK, so we have the same number of pixels as colors. That's a good thing. Now let's just pair pixels to colors.

```python
for px, c in Seq.zip pixels colors do 
    px.setRGB(c)
p
p.savePicture("out.png")
```

Let's "sort" the colors randomly and pair each color to a pixel.

```python
let random_colors =  colors |> Seq.sortBy (fun v -> Myro.random()) in
for px, c in Seq.zip pixels random_colors  do
    px.setRGB(c)
p
```

Rather than sorting the pixels, sort the pixels by the sum of their RGB value

```python
let sum_of_colors = colors |> Seq.sortBy (fun (r, g, b) -> r+g+b) in
for px, c in Seq.zip pixels sum_of_colors do 
    px.setRGB(c)
p
```

```python
let color_by_mag =  colors |> Seq.sortBy (fun (r, g, b) -> 
                                         (float r)**1.5 + (float g)**1.5 + (float b)**1.5) in
for px, c in Seq.zip pixels color_by_mag do
    px.setRGB(c)
p.savePicture("out.png")
p
```

```python
let pxs =  pixels |> Seq.sortBy (fun px -> px.x*px.x + px.y*px.y)
let cols = colors |> Seq.sortBy (fun (r, g, b) -> r + g + b)
for px, c in Seq.zip pxs colors do 
    px.setRGB(c)
p
```

```python
let cols = colors |> Seq.sortBy (fun (r, g, b) -> r*g)
for px, c in  Seq.zip pixels cols do
    px.setRGB(c)
p
```

```python
let almost_hue (r, g, b) =  (Math.Atan2(0.3**0.5 * (float (g - b)), (float (2*r - g - b))))
```

```python
let almost_hue_i (r, g, b) =  int (Math.Atan2(0.3**0.5 *(float (g - b)), (float (2*r - g - b))))
```

```python
let cols_by_hue = colors |> Seq.sortBy almost_hue
for px, c in  Seq.zip pixels cols_by_hue do
    px.setRGB(c)
p
```

```python
let cols_by_hue = colors |> Seq.sortBy almost_hue_i
for px, c in  Seq.zip pixels cols_by_hue do
    px.setRGB(c)
p
```

```python
let pxs =  pixels |> Seq.sortBy (fun px -> px.x*px.x + px.y*px.y)
let cols = colors |> Seq.sortBy (fun (r, g, b) -> almost_hue (r, g, b), r + g + b)
for px, c in Seq.zip pxs cols do 
    px.setRGB(c)
done
p
```
