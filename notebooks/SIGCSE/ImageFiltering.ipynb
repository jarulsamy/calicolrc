## Filtering with Convolution

```python
let filter_image(pic: Graphics.Picture) =  
    let filter =  array2D([ [0; -1; 0]; [-1; 4; -1];  [0; -1; 0]])
    let npic = Graphics.copyPicture(pic)
    for x in 1..pic.getWidth()-2 do
        for y in 1..pic.getHeight()-2 do
            let npx = npic.getPixel(x, y)
            let mutable sum = 0
            for i in -1..1 do
                for j in -1..1 do
                    sum <- sum + filter.[i+1,j+1] * pic.getPixel(x + i , y + j).getGray() 
                done
            done
            npx.setGray(sum)
        done
    done
    npic
```

### Python-Style Convolution (3x3 Kernel)

```python
let filter_image(pic: Graphics.Picture, filter: int [,]) =  
    let npic = Graphics.copyPicture(pic)
    for x in 1..pic.getWidth()-2 do
        for y in 1..pic.getHeight()-2 do
            let npx = npic.getPixel(x, y)
            let mutable sum = 0
            for i in -1..1 do
                for j in -1..1 do
                    let nx = x + i 
                    let ny = y + j
                    sum <- sum + filter.[i+1,j+1] * pic.getPixel(nx, ny).getGray() 
                done
            done
            npx.setGray(sum)
        done
    done
    npic
```

### Using Array2D.iteri to iterate over the kernel

```python
let filter_image(pic: Graphics.Picture, filter: int [,]) = 
    let npic = Graphics.copyPicture(pic)
    for x in 1..pic.getWidth()-2 do
        for y in 1..pic.getHeight()-2 do
            let npx = npic.getPixel(x, y)
            let sum = ref 0
            filter |> Array2D.iteri (fun i j v -> sum := !sum + v * pic.getPixel(x + i - 1, y + j - 1).getGray())
            npx.setGray(!sum)
        done
    done
    npic
```

### Create an Array2D.foldi

```python
module Array2D =
  let foldi f a matrix = 
    let mutable state = a
    for row in 0 .. Array2D.length1 matrix - 1 do
      for column in 0 .. Array2D.length2 matrix - 1 do
        state <- f row column (matrix.[row, column]) state
    state
```

```python
let filter_image(pic: Graphics.Picture, filter: int [,]) = 
    let npic = Graphics.copyPicture(pic)
    for x in 1..pic.getWidth()-2 do
        for y in 1..pic.getHeight()-2 do
            let npx = npic.getPixel(x, y)
            let mutable sum = 0
            sum <- filter |> Array2D.foldi (fun i j v s -> s + v * pic.getPixel(x + i - 1, y + j - 1).getGray()) 0
            if sum > 255 then sum <- 255
            if sum < 0 then sum <- 0
            npx.setGray(sum)
        done
    done
    npic
```

### Turn the pixels into an array before processing

```python
let filter_image(pic: Graphics.Picture, filter: int [,]) =     
    let pxarray = pic.getPixels() |> Seq.cast<Graphics.Pixel> |> Seq.map Graphics.getGray |> Array.ofSeq
    let w = pic.getWidth()
    let h = pic.getHeight()
    let kw = Array2D.length1 (filter)
    let kh = Array2D.length2 (filter)
    let npic = Graphics.copyPicture(pic)
    for x in 1..w-2 do
        for y in 1..h-2 do
            let npx = npic.getPixel(x, y)
            let mutable sum  = 0
            for i in 0..kw-1 do
                for j in 0..kh-1 do
                    let nx = x + i - kw/2
                    let ny = y + j - kh/2
                    let pos = ny + nx*h
                    //sum <- sum + filter.[i,j] * pic.getPixel(nx, ny).getGray() 
                    sum <- sum + filter.[i,j] * pxarray.[pos] 
                done
            done
            npx.setGray(sum)
        done
    done
    npic
```
