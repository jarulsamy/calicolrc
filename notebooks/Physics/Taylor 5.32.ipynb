## Problem restatement

First, we replicate the question. That helps in understanding exactly what is being asked, and to make sure we understand the equations:

**5.32 \*\*** [Computer] Consider an underdamped oscillator (such as a mass on the end of a spring) that is released from rest at position $x_o$ at time $t = 0$. 

**(a)** Find the position $x(t)$ at later times in the form

$$ x(t) = e ^{-\beta t}[B_1 cos(\omega_1 t) + B_2 sin(\omega_1 t)]. $$

That is, find $B_1$ and $B_2$ in terms of $x_0$.

**(b)** Now show that if you let $\beta$ approach the critical value $\omega_0$, your solution automaticall yields the critical solution.

**(c)** Using appropriate graphing software, plot the solution for $0 \leq t \leq 20$, with $x_0 = 1$, $\omega_0 = 1$, and $\beta = 0$, $0.02$, $0.1$, $0.3$, and $1$.

## Solution

### Matlab Solution

```matlab
t = 0:0.01:20;
b1 = 1;   % coefficient B1
omega0 = 1;
x0 = 1;
beta = [0, 0.02, 0.1, 0.3];
x = zeros(5,length(t));   % this array will hold the x(t) values for the 4 
                          % beta values other than beta = 1

lbls = char('beta = 0', 'beta = 0.02', 'beta = 0.1', 'beta = 0.3', 'beta = 1.0');
            % labels for the plot legend

% Loop through the beta values other than 1, accumulating the x(t) values for each  
         
for n = 1:4
    b = beta(n);     % temporary storage of current beta value
    omega1 = sqrt(omega0^2 - b^2);
    x(n,:) = exp(-b*t) .* (b1*cos(omega1*t) + (x0*b/omega1)*sin(omega1*t));
    plot(x(n,:));
    hold all;
end

% The x(t) function for beta = 1 has to be computed separately, since the
% solution is different.
xc = x0*exp(-1*t) + x0*t.*exp(-1*t);
plot(xc);
axis([0, 2000, -1, 1]);
xlabel('t');
ylabel('x(t)');
title('Underdamped Oscillation for Various Values of beta');
legend(lbls);
```

### Python Solution

First, we import the math library as we will need sin and cos:

````python
import math
````

Here, we focus on part (c). We are given $x_0$ and $\omega_0$. We set $B_1$ to 1, and compute $omega_1$ and $B_2$. Matching the equation above, we vary $t$, but also $\beta$:

````python
def x(t, beta):
    # Given:
    omega_0 = 1
    x_0 = 1
    if beta == 1:
        return x_0 * math.exp(-beta * t) + x_0 * t * math.exp(-beta * t)
    else:
        omega_1 = math.sqrt(omega_0 ** 2 - beta ** 2)
        B_1 = 1   
        B_2 = x_0 * beta/omega_1
        return math.exp(-beta * t) * ((B_1 * math.cos(omega_1 * t)) +  (B_2 * math.sin(omega_1 * t)))
````

Python doesn't have a floating-point range function (due to issues in rounding), so we code our own that will work for this problem:

````python
def frange(start, stop, step):
    indices = (i for i in range(0, int((stop-start)/step)))  
    return [start + step * i for i in indices]
````

We now test for a value of $\beta$:

````python
data = [x(t, beta=0) for t in frange(0, 20, .1)]
````

````python
calico.GoogleChart("LineChart", data)
````

Now, we compute the data for 4 values of $\beta$ (we save $\beta = 1$ as it takes a different form):

````python
all_data = []
for beta in [0, 0.02, 0.1, 0.3, 1]:
    all_data.append([x(t, beta) for t in frange(0, 20, .1)])
````

To make the data so that it groups all of values of $t$ together and pre-pends the actual value of $t$:

````python
lines = zip(frange(0, 20, .1), *all_data)
````

Finally, we plot all of values of $\beta$ in one plot:

````python
calico.GoogleChart("LineChart", 
                   ["time", "beta - 0", "beta - 0.02", "beta - 0.1", "beta - 0.3", "beta - 1.0"], 
                   lines)
````

## Interactive Plots

We can use the same function for interactively using Jupyter Interactive Widgets:

````python
beta = calico.FloatSliderWidget(description="beta", max=1, step=0.01)
def on_value_change(obj, value):
    chart = calico.GoogleChart("LineChart",
                               ["time", "beta - %s" % value],
                               [x(t, value) for t in frange(0, 20, .1)])
    calico.animate(chart)

beta.on_value_change(on_value_change, "value")
beta
````

## Animations

Finally, we can turn the same code into an animation:

````python
import Graphics

t = 0
while t < 20:
    canvas = Graphics.Canvas(100, 300)
    height = x(t, beta=0.1)
    spring = Graphics.Line((50, 0), (50, height * 150 + 150))
    spring.border = 5
    spring.draw(canvas)
    box = Graphics.Rectangle((50 - 20, height * 150 + 150 - 20), (50 + 20, height * 150 + 150 + 20))
    box.draw(canvas)
    calico.animate(canvas)
    t += 0.05
````
