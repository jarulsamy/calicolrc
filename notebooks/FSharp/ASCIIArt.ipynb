### Using ROS or the Camera Widget for ASCII Vision

As an homage to past days where high resolution displays and printers were rare, we will create a myro program that displays the robot's images using text characters rather than pixels. This is called [ASCII Art](http://en.wikipedia.org/wiki/Ascii_art). Let's write a simple program to create an ASCII image as a text file. You could also write the image as an HTML file to view in a web browser, but this example uses a simple text file.

```python
%%%lang fsharp
```

We nitialize our robot, and grab a color image from the camera. 

```python
Myro.makeRobot("ROSRobot", "192.168.1.2", "192.168.1.5")
```

We will save the picture to a jpeg file to later compare with our ASCII art masterpiece.

```python
let pic = Myro.takePicture()
pic
```

Or grab a picture from the webcam widget

```python
let w = calico.makeCameraWidget()
calico.display(w)
```

```python
let pic = Graphics.Picture(w.imageuri)
pic
```

Select an alphabet to use some ASCII art

```python
let alphabet = "@MHWRmEBSQKUGgqyp$8XDPFwdbkA&0ZTNhe9654YV*Cnsyza%3OLxo2JufIrc][vt}{71lji?|+)(=;-_!~:/,^.`"
```

```python
let pxs = Graphics.getPixels(pic)
let divider = 256 / alphabet.Length  + 1
let mutable out = ""
for j in 0..3..pic.getHeight()-1 do
     for i in 0..2..pic.getWidth()-1 do
         let pix = pic.getPixel(i, j)
         let intensity = pix.getGray()
         let v = intensity / divider
         out <- out + string(alphabet.[v])
     out <- out + "\n"
calico.animate(calico.HTML("<pre style=font-family:courier;font-size:8px;line-height:8px>" + out + "</pre>"))
```
