# Can a robot do something that it wasn't trained to do?

Let's do a simple robot learning experiment. First, let's start up a real robot or a simulated one: 

```python
import Myro
Myro.init("sim")
robot = Myro.getRobot()
Myro.setOption("show-sensors", True)
```

That should bring up a window like the following:

```python
Myro.getSimulation()
```

The infrared sensors (transparent blue pie pieces) work by hitting an obstacle, and bouncing back to the robot. The three on the front of the robot are much more sensitive than the ones on the back. The center-front sensor can be read by issuing <b>robot.getObstacle(1)</b>:

```python
robot.getObstacle(1)
```

The robot is too far from the wall. Let's move the robot forward at fullspeed (1), for one second (1). The command for that is <b>robot.forward(POWER, TIME)</b>:

```python
robot.forward(1, 1)
```

That will put the robot about right here (your milleage will vary... there is variation in the robot control and communication):

```python
Myro.getSimulation()
```

Now we check the reading again:

```python
robot.getObstacle(1)
```

Move a bit closer, and check again:

```python
robot.forward(1,1)
print(robot.getObstacle(1))
```

```python
Myro.getSimulation()
```

Interesting. The sensor reading gets larger the closer the robot gets to the wall. If we did further testing we would also find the up close, the reading is about 6400, and varies in a non-linear manner in relation to the distance.

We will use this single sensor in an attempt to let a neural network drive the robot. First, we will design a teacher program that will determine what the speed of the robot should be. As the robot gets closer to the wall, the teacher will instruct the robot to change speeds, using the <b>robot.forward(POWER)</b> command.

We need to create a neural network for the training.

```python
from ai.governor import GovernorNetwork
net = GovernorNetwork(2, 0.1, 0.1) # size of inputs + outputs
net.addLayers(1, 3, 1)
trial_count = 0
count = 0
```

Now, we start the training:

```python
robot.setPose(100, 250, 0)
net.setLearning(1)
t = 0

while True:
    count += 1
    target = 0.5
    obstacle = robot.getObstacle(1)
    distance = float(1 - obstacle/6400) 
    if distance < .5:
        target = 0.5
        if t == 0:
            robot.stop()
            t = Myro.currentTime()
        elif Myro.currentTime() - t >= 1.5:
            trial_count += 1
            robot.setPose(100, 250, 0)
            t = 0
        # else keep waiting
    elif distance < .8:
        target = .66
        robot.forward(.33)
    elif distance < .99:
        target = .83
        robot.forward(.66)
    else:
        target = 1.0
        robot.forward(1)
    results = net.step(input=[distance], output=[target])
    calico.animate("Trails: %s, Updates: %s" % (trial_count, count), str(net))
```

And let's test to see how well the neural network has learned what the teacher is doing.

```python
net.setLearning(0)
robot.setPose(100, 250, 0)
tdata = []
ndata = []
```

This is the teacher:

```python
def teacher(distance):
    if distance < .5:
        target = 0.5
    elif distance < .8:
        target = .66
    elif distance < .99:
        target = .83
    else:
        target = 1.0
    return target
```

Let's see what the network says to do in each of the positions:

```python
while True:
    obstacle = robot.getObstacle(1)
    distance = float(1 - obstacle/6400) 
    results = net.propagate(input=[distance])
    ndata.append(results[0])
    tdata.append(teacher(distance))
    speed = (results[0] * 2) - 1.0
    robot.forward(speed)
    calico.animate(speed)
```

Let's plot the teacher vs the network:

```python
from Widgets import Lines
size = len(tdata)
lines = Lines(size)
lines = Lines(lines, tdata, "Teacher")
lines = Lines(lines, ndata, "Network")
calico.display(calico.LineChart(lines))
```
