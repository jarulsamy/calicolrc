##Trees##

First, let's explore solutions organized in a '''tree'''. A ''tree'' is a recursive data structure which has three parts:

* an item, such as a value
* a left tree
* a right tree

As you can see, the definition of a tree contains two more trees, and thus makes it recursive. We can define a data structure in Python that exactly matches this definition:




If a tree doesn't have a left or right subtree, then we'll use the value <tt>None</tt>. You can make trees interactively:


```python
class Tree:
    def __init__(self, left, item, right):
        self.item = item
        self.left = left
        self.right = right

Tree(None, 42, None)
```

That creates a tree with a single element, or ''node''. We'll use the term node when we refer to a tree of just one element.

As a convenience, we'll add a method, called ''expand'' that will return the left and right nodes (if they exist) in a list:



```python
class Tree:
    def __init__(self, left, item, right):
        self.item = item
        self.left = left
        self.right = right

    def expand(self):
        retval = []
        if self.left:
            retval.append(self.left)
        if self.right:
            retval.append(self.right)
        return retval
```


That is a painful way to make trees. Let's create a function, called '''makeRandomTree()''' which takes a depth, and creates a tree that deep:


```python
from random import random

def makeRandomTree(n, min=0, max=1000):
    if n == 0: return None
    r = (max + min) // 2
    left = makeRandomTree(n - 1, min, r)
    right = makeRandomTree(n - 1, r, max)
    return Tree(left, r, right)
```

This will create a perfectly ''balanced'' tree of a depth ''n'':


```python
tree = makeRandomTree(4)
```

```python
def displayTree(tree):
    n = 0
    row = getRow(tree, n)
    while row:
        print(row)
        n += 1
        row = getRow(tree, n)
        
def getRow(tree, n, depth=0):
    if tree is None: 
        return []
    if depth == n: 
        return [tree.item]
    else:
        return getRow(tree.left, n, depth + 1) + getRow(tree.right, n, depth + 1)
        
displayTree(tree)
```

Also as a convenience, let's make a function called '''treeToList()''' that does exactly that:

```python
def treeToList(tree):
    retval = []
    if tree.left == None or tree.right == None:
        return tree.item
    retval.append(treeToList(tree.left))
    retval.append(tree.item)
    retval.append(treeToList(tree.right))
    return retval
```

## Stack and Queue ##

To search a Tree for items, we'll need two additional data structures, Stack, and Queue. A stack operates like a stack of plates in a cafeteria: the last one on the stack, is the first one off. A queue, on the other hand, is like a line to a movie: the first one in the queue is the first one into the show.

We'll give these two data structures identical methods:

* append(item) - adds an item to the queue/stack
* next() - removes and returns an item from the queue/stack
* is_empty() - returns True if the queue/stack has no more elements, and False otherwise



```python
class Stack:
    def __init__(self, start=None):
        self.list = [] if start == None else start

    def append(self, *items):
        for item in items:
            self.list.append(item)

    def next(self):
        return self.list.pop(-1)

    def is_empty(self):
        return len(self.list) == 0

class Queue(Stack):
    def next(self):
        return self.list.pop(0)
```

As Queue differs from Stack only in the manner that next() is implemented, we extend Queue from Stack.

## The Search is On ##

Now we can set up a general '''search()''' function. The function takes the items to search for in the form of a Stack or Queue of Trees.



```python
def search(items, match):
    while not items.is_empty():
        node = items.next()
        if node.item == match:
            return True
        items.append(*node.expand())
    return False
```

search uses the interface methods is_empty(), next(), and expand() so that it treats Stacks and Queues identically.

If we use Queue, then the nodes are visited (and this searched) in the order in which they were added: first-in, first-out. On the other hand, if we use Stack, then the nodes are visited (and this searched) in a first-in, last-out.

To make this easier to visualize what is occurring, we make some visualization tools.

First, we insert some graphics code into the search function:

```python
from Graphics import Graph, Color
import time

def search(g, items, match):
    while not items.is_empty():
        node = items.next()
        shape = g.getNode(str(node.item))
        shape.fill = Color("pink")
        calico.animate("Searching...", g.rerender())
        time.sleep(1)
        if node.item == match:
            return True
        items.append(*node.expand())
    return False
```

Now, we needs a graphics support function to reset the colors of a tree:

```python
def reset_color(g, tree):
    node = g.getNode(str(tree.item))
    node.fill = Color("white")
    if tree.left:
        reset_color(g, tree.left)
    if tree.right:
        reset_color(g, tree.right)
```

```python
calico.animate(g.rerender())
number = input("What number do you want to search for?")
while number:
    r = int(number)
    print(search(g, Queue([tree]), r))
    calico.animate("Searching for %s..." % r, g.rerender())
    reset_color(g, tree)
    number = input("What number do you want to search for?")
```


If we use Stack as the data structure for the search, then we have implemented a Breadth-first Search. On the other hand, if we use Queue as the data structure, then we have implemented a Depth-First Search.
