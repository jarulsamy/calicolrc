Getting a Camera image from your Computer
-----------------------------------------

1. Make a CameraWidget
2. display it
3. turn it into a Picture

```python
w = calico.makeCameraWidget()
```

```python
from Graphics import Picture
def f(value_name, value):
    calico.animate(Picture(w.imageuri))
    
w.on_value_change(f, "imageuri")
```

```python
calico.display(w)
```

```python
picture = Picture(w.imageuri)
```

```python
picture
```

```python
import Myro
for pixel in Myro.getPixels(picture):
    pixel.setRed(255)
```
